```python
categorical_variables=['gender','OnlineBackup','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod']
for var in categorical_variables:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(data[var], prefix=var)
    data1=data.join(cat_list)
    data=data1
```

```python
data1.drop(['Partner','Dependents','OnlineBackup','gender','PhoneService','MultipleLines','InternetService','OnlineSecurity','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling','PaymentMethod'], axis =1  , inplace =True )
```

```python
data1
```

```python
data1["TotalCharges"] = data1['TotalCharges'].replace(' ' , 0)
```

```python
data1
```

```python
X , y = data1.drop(['Churn', "customerID"], axis = 1 ) , data1['Churn']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test , y_train , y_test  = train_test_split(X  , y  ,test_size=0.3)
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train, y_train)
```

```python
pred = log_reg.predict(X_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
print(classification_report(pred , y_test))
```

```python
sns.heatmap(confusion_matrix(pred, y_test))
```
